```python
#File extraction and processing 
#extracting .O files, moves everything to an "old" folder
#here specify path 

#FIX- this is creating the data demo folder and then reading within it, rather than reading into the one with raw
# Define path patterns for zip files and output directories
pattern = {
    'MACROCOSM-5': 'C:/Users/cv273/Desktop/GNSSDat.2/data.demo/raw/MACROCOSM-5/MACROCOSM-5_raw_*.zip',
    'MACROCOSM-2': 'C:/Users/cv273/Desktop/GNSSDat.2/data.demo/raw/MACROCOSM-2/MACROCOSM-2_raw_*.zip',
    'MACROCOSM-1': 'C:/Users/cv273/Desktop/GNSSDat.2/data.demo/raw/MACROCOSM-1/MACROCOSM-1_raw_*.zip',
}

outputdir = {
    'MACROCOSM-5': 'data_demo/extracted/MACROCOSM-5/',
    'MACROCOSM-2': 'data_demo/extracted/MACROCOSM-2/',
    'MACROCOSM-1': 'data_demo/extracted/MACROCOSM-1/',
}

# Get list of zip files
filelists = gv.get_filelist(pattern)

# For each station, extract files and move zip files to "old" folder
for station_name, filelist in filelists.items():
    # Check if the output directory exists for that station
    if not os.path.exists(outputdir[station_name]):
        os.makedirs(outputdir[station_name])
    # Iterate through all files
    for izipfile in filelist:
        # Extract only the observation (.xxO) file in the outputdir
        with zipfile.ZipFile(izipfile, mode="r") as archive:
            for file in archive.namelist():
                if file.endswith("O"):
                    print(f"Extracting {file}")
                    archive.extract(file, outputdir[station_name])
        # Move the zip file to "old" folder
        old_folder = os.path.join(os.path.dirname(izipfile), "old")
        if not os.path.exists(old_folder):
            os.makedirs(old_folder)
        shutil.move(izipfile, old_folder)
```

```python
#calculate VOD

# set which data files should be loaded
pattern = 'data_demo/MACROCOSM_paired/*.nc'
# define how to associate stations together. Always put reference station first.
pairings = {'MACROCOSM':('MACROCOSM-5','MACROCOSM-2')}
# define if some observables with similar frequencies should be combined together. In the future, this should be replaced by the selection of frequency bands.
bands = {'VOD1':['S1','S1X','S1C'],'VOD2':['S2','S2X','S2C']}
#here change to utc5?
vod = gv.calc_vod(pattern,pairings,bands)
vod = vod['MACROCOSM']
vod
```

```python
import pytz
from datetime import datetime, timedelta

# Convert the MultiIndex into columns
vod_reset = vod.reset_index()

# Convert 'Epoch' column to datetime type
vod_reset['Epoch'] = pd.to_datetime(vod_reset['Epoch'])

# Define time zones
pr_tz = pytz.timezone('America/Puerto_Rico')
utc_tz = pytz.utc

# Convert time to Puerto Rico time zone
vod_reset['Epoch'] = vod_reset['Epoch'].dt.tz_localize(utc_tz).dt.tz_convert(pr_tz)

# Adjust for daylight saving time (DST)
# From first Sunday in April to last Sunday in October
dst_start = pytz.timezone('America/New_York').localize(datetime(2000, 4, 1))  # First Sunday in April
dst_end = pytz.timezone('America/New_York').localize(datetime(2000, 10, 31))   # Last Sunday in October
dst_range = pd.date_range(start=dst_start, end=dst_end, freq='D', tz='America/New_York')

# Apply the DST adjustment
adjusted_times = []
for epoch in vod_reset['Epoch']:
    if epoch.tz_localize(None) in dst_range:
        epoch = epoch - timedelta(hours=1)
    adjusted_times.append(epoch)

vod_reset['Epoch'] = adjusted_times

# Convert to UTC-4
vod_reset['Epoch'] = vod_reset['Epoch'].dt.tz_convert(utc_tz) - timedelta(hours=4)

# Set 'Epoch' column as the index
vod_reset.set_index('Epoch', inplace=True)
vod = vod_reset

print(vod)
```

```python
#average vod
# intialize hemispheric grid
hemi = gv.hemibuild(2)
# get patches for plotting later
patches = hemi.patches()
# classify vod into grid cells, drop azimuth and elevation afterwards as we don't need it anymore
vod = hemi.add_CellID(vod)
# get average value per grid cell
vod_avg = vod.groupby(['CellID']).agg(['mean', 'std', 'count'])
# flatten the columns
vod_avg.columns = ["_".join(x) for x in vod_avg.columns.to_flat_index()]
vod_avg
```
